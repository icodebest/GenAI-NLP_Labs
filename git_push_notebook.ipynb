```python
import os

def auto_push(commit_message="Auto-update from Colab"):
    """
    This function stages all changes, commits them with the provided message,
    and pushes them to the GitHub repository.
    """
    repo_dir = "/content/drive/MyDrive/MyProjectFolder"  # Update with your actual repo folder

    # Navigate to the repository
    os.chdir(repo_dir)

    # Add all changes
    os.system("git add .")

    # Commit changes
    os.system(f'git commit -m "{commit_message}"')

    # Push changes to GitHub
    os.system("git push origin main")  # Change 'main' if your branch name is different

    print("Changes pushed to GitHub successfully.")
```
